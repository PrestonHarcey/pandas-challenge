```python
import pandas as pd

school_data_upload = "../Resources/schools_complete.csv"
student_data_upload = "../Resources/students_complete.csv"


school_data = pd.read_csv(school_data_upload)
student_data = pd.read_csv(student_data_upload)
school_data_combo = pd.merge(student_data, school_data, on=["school_name", "school_name"])


print(school_data_combo)
```

```python
#creating Data Frames to hold the desired information
#create total_budget 
total_budget = school_data["budget"].sum()


#getting the totals for the schools and students.
#creating school_count using len() & .unique in order to count each school name once per instance
school_count = len(school_data_combo["school_name"].unique())
#create student_count using .count to account for repeat names that are different people
student_count = school_data_combo["Student ID"].count()


#Math

#calculate the average math score using .mean
average_math_score = school_data_combo["math_score"].mean()

#determine the passing score and the number of students who met or exceeded that score
passing_math_count = school_data_combo[(school_data_combo["math_score"] >= 70)].count()["student_name"]

#calculate percentage based percentage of total students graded at or above 70 (MULTIPLYING BY 100 for percentage!!)
passing_math_percentage = passing_math_count / float(student_count) * 100


#Reading

#calculate average reading score using .mean
average_reading_score = school_data_combo["reading_score"].mean()

#determine the passing score and the number of students who met or exceeded that score
passing_reading_count = school_data_combo[(school_data_combo["reading_score"] >= 70)].count()["student_name"]

#calculate percentage based percentage of total students graded at or above 70 (MULTIPLYING BY 100 for percentage!!)
passing_reading_percentage = passing_reading_count / float(student_count) * 100



passing_math_reading_count = school_data_combo[(school_data_combo["math_score"] >= 70) 
                                                  & (school_data_combo["reading_score"] >= 70)].count()["student_name"]

overall_passing_rate = passing_math_reading_count /  float(student_count) * 100 
#Note, thought it would be interesting to try  = school_data_combo["reading_score", "math_score"].mean() but dont know how
```

```python
#note: need to practice this section more. had issues with the formatting had to refer to example.


#this cell will create a DF using the information requested and the data provided to lable
District_Summary = pd.DataFrame({"Total Schools": [school_count], 
                                "Total Students": [student_count],
                                "Total Budget": [total_budget],
                                "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score],
                                "% Students Passing Math": [passing_math_percentage],
                                "% Students Passing Reading": [passing_reading_percentage],
                                "% Students Overall Passing": [overall_passing_rate]})

#note: this area was the hardest for me.
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)

#tried to add a title to my dataframe just for reference and did not have much luck looking around. maybe matplotlib?
District_Summary
```

```python
#start this df by determining it will be sorted by by the data type 'school'
School_Types = school_data.set_index(["school_name"])['type']

#using .value_counts() NOT VISCOUNT OR LEN to calculate students per school
per_school_counts = school_data_combo["school_name"].value_counts()
#create the budget values for calculation by grouping the budgets by name and find the average
per_school_budget = school_data_combo.groupby(["school_name"]).mean()["budget"]
#divide budged by counts in order to determine the average budget per student
per_school_capita = per_school_budget / per_school_counts

# Get the students who passed math and passed reading by creating separate filtered DataFrames.Almost messed this one up bad
school_passing_math = school_data_combo[(school_data_combo["math_score"] >= 70)]
school_passing_reading = school_data_combo[(school_data_combo["reading_score"] >= 70)]

# Creating a df specifically to hold the students who passed both math AND(&) Reading
passing_math_and_reading = school_data_combo[(school_data_combo["reading_score"] >= 70)
                                               & (school_data_combo["math_score"] >= 70)]

#obtain values for average math score per school
per_school_math = school_data_combo.groupby(["school_name"]).mean()["math_score"]
#obtain the amount of students passing math per school (MULTIPLYING BY 100 for percentage!!)
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

#obtain values for average reading score per school
per_school_reading = school_data_combo.groupby(["school_name"]).mean()["reading_score"]
#obtain the amount of students passing reading per school(MULTIPLYING BY 100 for percentage!!)
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

#overall rate per school using passingmathandreading from earlier (MULTIPLYING BY 100 for percentage!!)
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

print(overall_passing_rate)
```

```python
#Creating the School Summary DF
per_school_summary = pd.DataFrame({"School Type": School_Types,
                                  "Total Students": per_school_counts,
                                  "Total School Budget": per_school_budget,
                                  "Per Student Budget": per_school_budget,
                                  "Average Math Score": per_school_math,
                                  "Average Reading Score": per_school_reading,
                                   Math": per_school_passing_math,
                                   Reading": per_school_passing_reading,
                                  "% Students Overall Passing": overall_passing_rate})


per_school_summary
```

```python
#simple df using a sorting of the schools by percentage overall passing in order to display the top performing schools
top_schools = per_school_summary.sort_values(["% Overall Passing"], ascending=False)
top_schools.head(5)
```

```python
#a simple df holding the bottom schools performance by overall passing
bottom_schools = per_school_summary.sort_values(["% Overall Passing"], ascending=True)
bottom_schools.head(5)
```

```python
#Create the Scores by Grade DF
#conditional formatting to grab values per grade
ninth_graders = school_data_combo[(school_data_combo["grade"] == "9th")]
tenth_graders = school_data_combo[(school_data_combo["grade"] == "10th")]
eleventh_graders = school_data_combo[(school_data_combo["grade"] == "11th")]
twelfth_graders = school_data_combo[(school_data_combo["grade"] == "12th")]

#grouping methods to group each result per grade to the cooresponding school
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]

#combining both of those metrics into one DF
scores_by_grade = pd.DataFrame({"9th": ninth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})


scores_by_grade
```

```python
#Scores by school spending
#note: need to revisit BIN videos to relearn how they interact, misremembered information
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
#creating a copy is a much easier way than trying to link the original dataset into this df, remember for future use
school_spending_df = per_school_summary

#organize the data using the bins created
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)


```

```python
#using the newly made school spend df I can take the average of the new values from the scores to create new data

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

#Creating the Dataframe Spending Summary
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores,
                                "Average Reading Score": spending_reading_scores,
                                "% Students Passing Math": spending_passing_math,
                                "% Students Passing Reading": spending_passing_reading,
                                "% Students Overall Passing": overall_passing_spending})
spending_summary
```
